# Messing around with numpy and matplotlib

```python
# Create plots using pyplot from matplotlib

from matplotlib import pyplot as plt
import numpy as np
```

```python
# Create 1D arrays using np.arange for data values

x = np.arange(0, 5, 1)
y = np.arange(0, 10, 2)
z = np.arange(0, -10, -2)
exp = x * x
print(x, y, z, exp)
```

```python
# Style and plot data 

plt.axhline(color = 'black', linewidth = 0.5)
pos = plt.plot(x, y, marker = 'o', label = 'x vs. y')
neg = plt.plot(x, z, marker = 'o', label = 'x vs. z')
exp = plt.plot(x, exp, label = 'exponential')
plt.legend(loc = 'upper left')
plt.show()
```
